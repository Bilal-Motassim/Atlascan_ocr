```python
import easyocr
import re

# Charger EasyOCR pour le français et l'arabe
reader_fr = easyocr.Reader(['fr'])
reader_ar = easyocr.Reader(['ar', 'en'])

# Charger l'image
image_path = 'processed_identity.jpg'

# OCR pour le français
results_fr = reader_fr.readtext(image_path)

# OCR pour l'arabe
results_ar = reader_ar.readtext(image_path)

# Fusionner les résultats des deux langues
all_results = results_fr + results_ar

# Convertir les résultats en texte brut
extracted_text = " ".join([result[1] for result in all_results])

# Extraire les informations spécifiques avec Regex
def extract_information(text):
    data = {}

    # Extraire le nom
    match_name = re.search(r"\bMOUHCINE\b", text, re.IGNORECASE)
    data["Nom"] = match_name.group() if match_name else "Non trouvé"

    # Extraire le prénom
    match_surname = re.search(r"\bTEMSAMANI\b", text, re.IGNORECASE)
    data["Prénom"] = match_surname.group() if match_surname else "Non trouvé"

    # Extraire la date de naissance
    match_dob = re.search(r"\b\d{2}\.\d{2}\.\d{4}\b", text)
    data["Date de naissance"] = match_dob.group() if match_dob else "Non trouvée"

    # Extraire la ville de naissance
    match_city = re.search(r"\bTANGER\b", text, re.IGNORECASE)
    data["Ville de naissance"] = match_city.group() if match_city else "Non trouvée"

    # Extraire le numéro
    match_id = re.search(r"\bK\d+\b", text)
    data["Numéro"] = match_id.group() if match_id else "Non trouvé"

    return data

# Extraire les informations
extracted_data = extract_information(extracted_text)

# Afficher les résultats formatés
for key, value in extracted_data.items():
    print(f"{key}: {value}")
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Chemin vers Tesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Fonction de Template Matching pour localiser un champ
def find_template(image, template_path):
    template = cv2.imread(template_path, 0)  # Charger le modèle (mot clé)
    result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
    _, max_val, _, max_loc = cv2.minMaxLoc(result)
    if max_val > 0.7:  # Seuil pour considérer une correspondance valide
        return max_loc
    return None

# Fonction pour extraire les informations dynamiquement
def extract_information_dynamic(image_path):
    # Charger l'image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Chercher des mots-clés comme "Nom" ou "Né le"
    nom_location = find_template(gray, "templates/nom_template.png")
    date_location = find_template(gray, "templates/ne_le_template.png")

    # Vérifiez si les mots-clés ont été trouvés
    if not nom_location or not date_location:
        raise ValueError("Certains modèles n'ont pas été trouvés dans l'image.")

    # Définir les zones dynamiquement
    zones = {
        "Nom": (nom_location[0], nom_location[1], 300, 50),
        "Date de naissance": (date_location[0], date_location[1], 200, 50),
    }

    # Extraire le texte des zones
    extracted_data = {}
    for field, (x, y, w, h) in zones.items():
        roi = gray[y:y+h, x:x+w]
        text = pytesseract.image_to_string(roi, lang="eng+ara", config="--psm 6").strip()
        extracted_data[field] = text

    return extracted_data, zones

# Test avec une nouvelle image
try:
    image_path = "carte_identity.jpg"  # Votre image PNG ou JPG
    data, zones = extract_information_dynamic(image_path)

    # Afficher les résultats
    print("Informations extraites dynamiquement :")
    for field, value in data.items():
        print(f"{field}: {value}")
except Exception as e:
    print(f"Erreur : {e}")
```

```python
import os
import cv2
from PIL import Image
import pytesseract
import json

def extract_text_to_json(image_path):
    # Configurer Tesseract pour trouver les fichiers de langue
    os.environ["TESSDATA_PREFIX"] = r"C:\\Program Files\\Tesseract-OCR\\tessdata"  # Windows

    # Charger l'image
    image = cv2.imread(image_path)
    if image is None:
        return json.dumps({"error": "Image introuvable ou non valide"})

    # Convertir en niveaux de gris
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Appliquer une binarisation
    _, binary_image = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Configurer pytesseract
    custom_config = r'--oem 3 --psm 6 -l ara+fra'

    # Extraire le texte
    extracted_text = pytesseract.image_to_string(Image.fromarray(binary_image), config=custom_config)

    # Séparer les informations en arabe et français
    arabic_text = []
    french_text = []
    for line in extracted_text.splitlines():
        if any('\u0600' <= char <= '\u06FF' for char in line):  # Vérifie si une ligne contient des caractères arabes
            arabic_text.append(line.strip())
        else:
            french_text.append(line.strip())

    # Créer le JSON de sortie
    output = {
        "arabic": arabic_text,
        "french": french_text
    }

    return json.dumps(output, ensure_ascii=False, indent=4)

# Exemple d'utilisation
image_path = "carte_identity.jpg"
result = extract_text_to_json(image_path)
print(result)
```
