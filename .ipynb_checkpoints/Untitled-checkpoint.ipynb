```python
import easyocr
import re

# Charger EasyOCR pour le français et l'arabe
reader_fr = easyocr.Reader(['fr'])
reader_ar = easyocr.Reader(['ar', 'en'])

# Charger l'image
image_path = 'processed_identity.jpg'

# OCR pour le français
results_fr = reader_fr.readtext(image_path)

# OCR pour l'arabe
results_ar = reader_ar.readtext(image_path)

# Fusionner les résultats des deux langues
all_results = results_fr + results_ar

# Convertir les résultats en texte brut
extracted_text = " ".join([result[1] for result in all_results])

# Extraire les informations spécifiques avec Regex
def extract_information(text):
    data = {}

    # Extraire le nom
    match_name = re.search(r"\bMOUHCINE\b", text, re.IGNORECASE)
    data["Nom"] = match_name.group() if match_name else "Non trouvé"

    # Extraire le prénom
    match_surname = re.search(r"\bTEMSAMANI\b", text, re.IGNORECASE)
    data["Prénom"] = match_surname.group() if match_surname else "Non trouvé"

    # Extraire la date de naissance
    match_dob = re.search(r"\b\d{2}\.\d{2}\.\d{4}\b", text)
    data["Date de naissance"] = match_dob.group() if match_dob else "Non trouvée"

    # Extraire la ville de naissance
    match_city = re.search(r"\bTANGER\b", text, re.IGNORECASE)
    data["Ville de naissance"] = match_city.group() if match_city else "Non trouvée"

    # Extraire le numéro
    match_id = re.search(r"\bK\d+\b", text)
    data["Numéro"] = match_id.group() if match_id else "Non trouvé"

    return data

# Extraire les informations
extracted_data = extract_information(extracted_text)

# Afficher les résultats formatés
for key, value in extracted_data.items():
    print(f"{key}: {value}")
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Chemin vers l'exécutable Tesseract (Windows uniquement)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Dimensions de référence basées sur l'image testée
REFERENCE_WIDTH = 1000
REFERENCE_HEIGHT = 800

# Zones définies pour l'image de référence
zones = {
    "Nom": (440, 210, 280, 50),
    "Prénom": (440, 292, 280, 50),
    "Date de naissance": (740, 340, 200, 50),
    "Ville de naissance": (460, 430, 500, 50),
    "Numéro": (170, 730, 230, 50),
}

# Fonction pour redimensionner une image à la taille de référence
def normalize_image(image, target_width, target_height):
    return cv2.resize(image, (target_width, target_height))

# Fonction pour ajuster dynamiquement les zones en fonction de la taille de l'image
def adjust_zones(zones, original_width, original_height, target_width, target_height):
    scale_x = target_width / original_width
    scale_y = target_height / original_height
    adjusted_zones = {
        field: (
            int(x * scale_x),
            int(y * scale_y),
            int(w * scale_x),
            int(h * scale_y)
        )
        for field, (x, y, w, h) in zones.items()
    }
    return adjusted_zones

# Fonction pour afficher l'image avec les zones
def visualize_zones(image, zones):
    for field, (x, y, w, h) in zones.items():
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, field, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 6))
    plt.imshow(image_rgb)
    plt.axis("off")
    plt.title("Zones définies")
    plt.show()

# Fonction pour extraire les informations d'une image
def extract_information(image_path):
    # Charger l'image
    image = cv2.imread(image_path)
    original_height, original_width = image.shape[:2]

    # Ajuster dynamiquement les zones en fonction des dimensions de l'image
    adjusted_zones = adjust_zones(zones, original_width, original_height, REFERENCE_WIDTH, REFERENCE_HEIGHT)

    # Normaliser l'image à la taille cible
    normalized_image = normalize_image(image, REFERENCE_WIDTH, REFERENCE_HEIGHT)

    # Prétraiter l'image (grayscale et binarisation)
    gray = cv2.cvtColor(normalized_image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # Fonction pour extraire le texte d'une zone
    def extract_text_from_zone(image, zone, lang="eng+ara"):
        x, y, w, h = zone
        roi = image[y:y+h, x:x+w]  # Découpe de la région d'intérêt
        text = pytesseract.image_to_string(roi, lang=lang, config="--psm 6").strip()
        return text

    # Extraire les données pour chaque champ
    extracted_data = {}
    for field_name, coordinates in adjusted_zones.items():
        extracted_data[field_name] = extract_text_from_zone(binary, coordinates)

    return extracted_data, normalized_image, adjusted_zones

# Appliquer la fonction pour une nouvelle image
image_path = "carte_identity.jpg"  # Chemin de la nouvelle image
data, normalized_image, adjusted_zones = extract_information(image_path)

# Visualiser les zones ajustées sur l'image normalisée
visualize_zones(normalized_image, adjusted_zones)

# Afficher les résultats extraits
print("Informations extraites :")
for field, value in data.items():
    print(f"{field}: {value}")
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Chemin vers Tesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Dimensions de référence
REFERENCE_WIDTH = 1000
REFERENCE_HEIGHT = 800

# Zones définies pour une image de référence
zones = {
    "Nom": (189, 70, 80, 30),
    "Prénom": (189, 110, 70, 15),
    "Ville de naissance": (189, 159, 120, 30),
     "Date de naissance": (460, 430, 500, 50),
    "Numéro": (170, 730, 230, 50),
}

# Fonction pour redimensionner une image à la taille cible
def normalize_image(image, target_width, target_height):
    return cv2.resize(image, (target_width, target_height))

# Ajuster les zones dynamiquement
def adjust_zones_to_image(zones, original_width, original_height, target_width, target_height):
    scale_x = target_width / original_width
    scale_y = target_height / original_height
    adjusted_zones = {}
    for field, (x, y, w, h) in zones.items():
        new_x = int(x * scale_x)
        new_y = int(y * scale_y)
        new_w = int(w * scale_x)
        new_h = int(h * scale_y)

        # Vérifiez que les zones ne dépassent pas les dimensions de l'image
        if new_x + new_w > target_width or new_y + new_h > target_height:
            print(f"Avertissement : La zone '{field}' dépasse les dimensions ajustées. Elle sera ignorée.")
            continue
        adjusted_zones[field] = (new_x, new_y, new_w, new_h)
    return adjusted_zones

# Fonction pour afficher les zones
def visualize_zones(image, zones):
    for field, (x, y, w, h) in zones.items():
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, field, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 6))
    plt.imshow(image_rgb)
    plt.axis("off")
    plt.title("Zones définies")
    plt.show()

# Fonction pour extraire les informations
def extract_information(image_path):
    # Charger l'image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if image is None:
        raise ValueError(f"L'image n'a pas pu être chargée. Vérifiez le chemin : {image_path}")

    original_height, original_width = image.shape[:2]

    # Ajuster dynamiquement les zones
    adjusted_zones = adjust_zones_to_image(zones, original_width, original_height, REFERENCE_WIDTH, REFERENCE_HEIGHT)

    # Redimensionner l'image à la taille cible
    normalized_image = normalize_image(image, REFERENCE_WIDTH, REFERENCE_HEIGHT)

    # Prétraiter (grayscale et binarisation)
    gray = cv2.cvtColor(normalized_image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # Fonction pour extraire le texte d'une zone
    def extract_text_from_zone(image, zone, lang="eng+ara"):
        x, y, w, h = zone
        if x + w > image.shape[1] or y + h > image.shape[0]:  # Vérification supplémentaire
            print(f"Avertissement : Zone hors limites pour {zone}.")
            return ""
        roi = image[y:y+h, x:x+w]
        text = pytesseract.image_to_string(roi, lang=lang, config="--psm 6").strip()
        return text

    # Extraire les données pour chaque champ
    extracted_data = {}
    for field_name, coordinates in adjusted_zones.items():
        extracted_data[field_name] = extract_text_from_zone(binary, coordinates)

    return extracted_data, normalized_image, adjusted_zones

# Test avec une nouvelle image
image_path = "carte.png"  # Remplacez par le chemin de votre image
try:
    data, normalized_image, adjusted_zones = extract_information(image_path)

    # Visualiser les zones
    visualize_zones(normalized_image, adjusted_zones)

    # Afficher les résultats
    print("Informations extraites :")
    for field, value in data.items():
        print(f"{field}: {value}")
except Exception as e:
    print(f"Erreur : {e}")
```
